```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
cars = pd.read_csv('C:/Users\moham\Jupyter Practice\Exercise Files\Data\mtcars.csv')
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'wsec', 'vs', 'am', 'gear', 'carb']

cars.index = cars.car_names

mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

#Seaborn Histogram

```python
sb.distplot(mpg)
```

# Scatterplots

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgrey'], s=150)
```

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

Scatter plot matrix

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
```

Building boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
