```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

```python
x = range(1, 10)
y = [1, 2, 3, 4, 5, 4, 3, 2, 1]

plt.bar(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
z = [1, 2, 3, 4, 0.5]

veh_type = ['Bicycle', 'MotorBike', 'Car', 'Van', 'Stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

```python
cars = pd.read_csv('C:/Users\moham\Jupyter Practice\Exercise Files\Data\mtcars.csv')
cars.columns = ['car_names', 'mpg', 'cylinders', 'displacement', 'hp', 'drat', 'wt', 'wsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')
```
